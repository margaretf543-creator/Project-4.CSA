```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#histogram car prices 
df_filtered = df[df['price']<100000]
fig = px.histogram(df, x='price', title="Distribution of Car Prices") 
fig.show()
```

```python
#histogram of odometer
fig = px.histogram(df, x='odometer', title='Distribution of Odometer Readings')
fig.show()
```

```python
#scatter plot: price vs model year:
fig = px.scatter(df, x='model_year', y='price', title='Price vs Model Year', color='condition')
fig.show()
```

```python
#scatter plot: price vs odometer 
fig = px.scatter(df, x='odometer', y='price', title='Price vs Odometer Reading', color='condition')
fig.show()
```

```python
#histogram by condition
fig= px.histogram(df, x='condition', title='Number of cars by condition')
fig.show()
```
